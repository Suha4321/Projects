```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# for jupyter to render the graph
%matplotlib inline
```

```python
graduate = pd.read_csv('/Users/Suha/Projects/TestGitHub/WomenBachelors/percent-bachelors-degrees-women-usa.csv')
```

## Generate a line chart that visualizes the historical percentage of Biology degrees awarded to women and compare it to that awared to men

```python
fig = plt.figure(figsize = (10,10))
                 
    
ax1 = fig.add_subplot(2,1,1)
ax1.plot(graduate['Year'], graduate['Biology'], c = 'b')
ax1.plot(graduate['Year'], 100 - graduate['Biology'], c = 'r')
ax1.set_xlabel("Year")
ax1.set_ylabel("% Graduation")
ax1.legend(["Women", "Men"], loc = 'upper right')
ax1.set_title("Percentage of Biology Degrees Awarded By Gender")
```

## now making some improvement based on the data- ink principle
1) Remove all of the axis tick marks.
2) Hide the spines, which are the lines that connects the tick marks, on each axis.

```python
fig = plt.figure(figsize = (10,10))
                 
    
ax2 = fig.add_subplot(2,1,1)
ax2.plot(graduate['Year'], graduate['Biology'], c = 'b')
ax2.plot(graduate['Year'], 100 - graduate['Biology'], c = 'r')
ax2.set_xlabel("Year")
ax2.set_ylabel("% Graduation")
ax2.legend(["Women", "Men"], loc = 'upper right')
ax2.set_title("Percentage of Biology Degrees Awarded By Gender")

# Remove all of the axis tick marks.
ax2.tick_params(top = 'off', bottom = 'off', left = 'Off', right = 'off')

# Hide the spines, which are the lines that connects the tick marks, on each axis.
print(ax2.spines)
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax2.spines['bottom'].set_visible(False)
ax2.spines['left'].set_visible(False)
```

## Let's generate line charts for four STEM degree categories on a grid to encourage comparison.

```python
fig = plt.figure(figsize = (20,20))
                 
subjects = ['Biology', 'Engineering', 'Computer Science','Math and Statistics' ]    


for i in range(0,4):
#     print((4,plot_column[i],plot_place[i]))
    ax = fig.add_subplot(2,2,1+i)
    ax.plot(graduate['Year'], graduate[subjects[i]], c = 'b', label = 'Women')
    ax.plot(graduate['Year'], 100 - graduate[subjects[i]], c = 'g', label = 'Men')
    ax.tick_params(top = 'off', bottom = 'off', left = 'Off', right = 'off')
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.set_title(subjects[i])
    
ax.legend(['Women','Men'], loc = 'upper right')
```

## we can say that just based on the graduation data as a proxy for gender gap in STEM, the enineering and computer science fields have a much wider gap as compared to Math Statistics and Biology field.

## Let us go ahead and customize colors, line widths, layout, and annotations to improve the ability for a viewer to extract insights from the charts.
For this we will first build our color palette    

```python
#Color blind 10 palette
blue1 = (0/255,107/255,164/255)
orange1 = (255/255,128/255,14/255)
grey2 = (171/255,171/255,171/255)
dark_grey = (89/255,89/255,89/255)
blue2 = (95/255,158/255,209/255)
brick_orange = (200/255,82/255,0/255)
grey1 = (137/255,137/255,137/255)
blue3 = (162/255,200/255,236/255)
orange2 = (255/255,188/255,121/255)
grey3 = (207/255,207/255,207/255)
```

```python
# changing the line color to suit the color blind person requirement
# Changeing the line width to 3
fig = plt.figure(figsize = (12,12))
                 
subjects = ['Biology', 'Engineering', 'Computer Science','Math and Statistics' ]    


for i in range(0,4):
#     print((4,plot_column[i],plot_place[i]))
    ax = fig.add_subplot(2,2,1+i)
    ax.plot(graduate['Year'], graduate[subjects[i]], c = blue1, label = 'Women', linewidth = 2)
    ax.plot(graduate['Year'], 100 - graduate[subjects[i]], c = orange1, label = 'Men', linewidth = 2)
    ax.tick_params(top = 'off', bottom = 'off', left = 'Off', right = 'off')
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.set_title(subjects[i])
    ax.set_xlim(1968,2011)
    
ax.legend(['Women','Men'], loc = 'upper right')

```

Let us go ahead and visualise the gender gap for all the 6 subjects. For making the visuals coherent we will order the graphs in one row and wiht decreasing gender gap

```python
fig = plt.figure(figsize = (24,4))
                 
subjects = ['Engineering', 'Computer Science','Psychology', 'Biology', 'Physical Sciences','Math and Statistics' ]    



for i in range(0,5):
#     print((4,plot_column[i],plot_place[i]))
    ax = fig.add_subplot(1,6,1+i)
    ax.plot(graduate['Year'], graduate[subjects[i]], c = blue1, label = 'Women', linewidth = 2)
    ax.plot(graduate['Year'], 100 - graduate[subjects[i]], c = orange1, label = 'Men', linewidth = 2)
    ax.tick_params(top = 'off', bottom = 'off', left = 'Off', right = 'off')
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_title(subjects[i])
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    if i ==0:
        ax.text(2005,87,'Men')
        ax.text(2002,8,'Women')
    if i ==5:
        ax.text(2005,62, 'Men')
        ax.text(2001,35,'Women')


    
# ax.legend(['Women','Men'], loc = 'upper right')

```
